## Searborn for Data Visualization

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

### Built in Datasets in Seaborn

```python
penguins= sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.dropna(inplace=True)
penguins.head()
```

### Scatterplot

```python
sns.set_style('ticks')
sns.set_context('paper')
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='island', style='sex', palette='Set1')
sns.despine(top=True,right=True)
```

```python
sns.set_style('ticks')
sns.set_context('notebook')
sns.scatterplot(data=penguins, x='species', y='body_mass_g', hue='island', style='sex', palette='Set1', alpha=0.5)
sns.despine(top=True,right=True)
```

### Stripplot

```python
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(data=penguins, x='species', y='body_mass_g', hue='island', palette='Set1', jitter=True, dodge=True)
sns.despine(top=True, right=True)
```

### Swarmplot

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
sns.swarmplot(data=penguins, x='species', y='body_mass_g', hue='island', palette='Set1', dodge=True)
sns.despine(top=True, right=True)
```

### Histrogram

```python
sns.histplot(data=penguins, x='body_mass_g', hue='sex', multiple='stack', palette='Dark2')
```
