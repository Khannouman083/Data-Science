## Data Visualization Using Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[1,2,3]
y=[4,5,6]

plt.plot(x,y)
plt.show()
```

### Univariate (Single Column Numerical)

```python
data = {
    "Salary" : [50000, 60000, 70000, 80000, 110000, 115000, 150000]
}
df= pd.DataFrame(data)
df
```

### Line Plot

```python
plt.plot(df["Salary"], color = 'red', marker = 'o', linestyle= "--", label = "Salary", linewidth = 2)
plt.grid()
plt.legend()
plt.ylabel("Salary is Rs.")
plt.show()
```

### Histogram (Frequency Distribution of Data)

```python
plt.hist(df["Salary"], bins=7, color='green', edgecolor = 'black', label="Salary Distribution")
plt.legend()
plt.show()
```

### Box Plot

```python
plt.boxplot(df["Salary"], vert=True,patch_artist=True, tick_labels=['Salary'], )
plt.show()
```

```python
df.loc[7] = 1500000
plt.boxplot(df["Salary"], vert=True, patch_artist=True,tick_labels=['Salary'], )
plt.show()
```

```python
df.drop(index=7, inplace=True)
plt.boxplot(df["Salary"], vert=True, patch_artist=True,tick_labels=['Salary'], )
plt.show()
```

## Univartiate (Categorical Single Column)

```python
df['Dept'] = ['HR', 'Sales', 'IT', 'HR', 'Sales', 'HR', 'IT']
count = df['Dept'].value_counts()
```

### Pie Chart

```python
plt.pie(count, labels=count.index, autopct = "%1.1f", colors=['red', 'blue', "green"], explode=[0,0.1,0])
plt.axis("equal")
plt.show()
```

### Count Plot

```python
plt.bar(count.index, count, color=['red', 'blue', 'green'], edgecolor='black')
plt.show()
```
