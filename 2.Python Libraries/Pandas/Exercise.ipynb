```python
import numpy as np
import pandas as pd
```

```python
try:
    data = pd.read_csv(r"D:\Python\Libraries\Pandas\sales_data.csv", encoding='latin1')
except FileNotFoundError:
    print("File Not Found")
data
```

```python
data.info()
```

```python
data["Date"]= pd.to_datetime(data["Date"])
data.info()
```

```python
data["Units Sold"]=pd.to_numeric(data["Units Sold"])
data["Unit Price"]=pd.to_numeric(data["Unit Price"])
```

```python
data.isnull().sum()
data.fillna(0)
```

```python
data["Total Sales"] = data["Unit Price"] * data["Units Sold"]
data
```

```python
total_sales_by_stores=data.groupby("Store")["Total Sales"].sum().reset_index()
print(total_sales_by_stores)
```

```python
best_selling_product = data.groupby("Product")["Total Sales"].sum().reset_index()
print(best_selling_product.head(1))
```

```python
sales_per_day = data.groupby("Date")["Total Sales"].sum().reset_index()
total_sales_day = sales_per_day.sort_values(by="Total Sales", ascending=False)
print(total_sales_day.head(1))
```

```python
units_per_day=data.groupby("Date")["Units Sold"].sum()
average_units =np.mean(units_per_day)
print(average_units)
```
