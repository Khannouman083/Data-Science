## NumPy Operations on Real Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
  #S.no  2019  2020  2021  2022  2023  2024
    [1, 2500, 2700, 2900, 3000, 3200, 3500],  # Restaurant 1
    [2, 1800, 2000, 2300, 2600, 2900, 3100],  # Restaurant 2
    [3, 3000, 3100, 3300, 3600, 3900, 4200],  # Restaurant 3
    [4, 2200, 2400, 2500, 2700, 2800, 3000],  # Restaurant 4
    [5, 1500, 1700, 1900, 2000, 2300, 2600]   # Restaurant 5
])
```

### Printing or accessing Specific Data

```python
print(sales_data[1:5]) #print all rows except the first
print(sales_data[:,1:]) #print all columns except the first column
```

### Total Sales Each Restaurent in all years

```python
restaurent_sales=np.sum(sales_data[:,1:],axis=1)
print(restaurent_sales)
```

### Totals Sales for each year

```python
yearly_sales=np.sum(sales_data[:,1:], axis=0)
print(yearly_sales)
```

### Minimum and Maximum Sales per Restaurent

```python
min_rest_sales=np.min(sales_data[:,1:], axis=1)
print(min_rest_sales)

max_rest_sales=np.max(sales_data[:,1:], axis=1)
print(max_rest_sales)
```

### Minimum and Maximum Sales Per Year

```python
min_yearly_sales=np.min(sales_data[:,1:], axis=0)
print(min_yearly_sales)

max_yearly_sales=np.max(sales_data[:,1:], axis=0)
print(max_yearly_sales)
```

### Average Sales Per Year and Restaurent

```python
avg_yearly_sales=np.mean(sales_data[:,1:], axis=0)
print(avg_yearly_sales)

avg_rest_sales=np.mean(sales_data[:,1:], axis=1)
print(avg_rest_sales)
```

### Cumulative Sums Per Year

```python
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)
```

### Printing the data using Matplotlib

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cumulative Sum Per Year")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show
```

### Vector Operations

```python
arr1=np.array([[1,2],[3,4],[5,6]])
arr2=np.array([[2,2],[5,4],[5,7]])
arr3=np.array([[1,2,3],[4,5,6]])

print("Sum: ",arr1+arr2)
print("Mult: ",arr1*arr2)
print("Dot Product: ",np.dot(arr1,arr3))
```

### Angle Between Vectors

```python
vector1=np.array([1,2,3])
vector2=np.array([3,4,5])
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```

### Vectroize Operations

```python
str_arr = np.array(["Nouman", "Khan", "Hello"])
vectorize_upper = np.vectorize(str.upper)
vectroize_lower = np.vectorize(str.lower)

print(vectorize_upper(str_arr))
print(vectroize_lower(str_arr))
```

### Broadcast

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])
print(A + b)
```
