## More On Plots in Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np

penguins= sns.load_dataset('penguins')
penguins.head()
```

### Violin Plot

```python
sns.violinplot(data=penguins, x='species', y='body_mass_g', hue='island', split=False)
```

### Kde Plot

```python
sns.kdeplot(data=penguins, x='body_mass_g', hue='species', fill=True)
```

### Heatmap

```python
columns = ["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]
sns.heatmap(data=penguins[columns].corr(), annot=True, cmap='coolwarm', vmin=-0.2, vmax=0.8)
```

### Rug Plot

```python
sns.rugplot(data=penguins, y='body_mass_g', hue='species', height=0.5)
```

### Pair Plot

```python
sns.pairplot(data=penguins, hue='species', diag_kind='hist', kind='scatter')
```

### Pair Grid

```python
graph = sns.PairGrid(data=penguins, hue='species')
graph.map_diag(sns.histplot)
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
```
