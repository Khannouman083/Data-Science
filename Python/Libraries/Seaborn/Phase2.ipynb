## Some Other Types of Plots

```python
import seaborn as sns
import pandas as pd
import numpy as np

penguins= sns.load_dataset('penguins')
penguins.head()
```

### Regression Plot

```python
sns.regplot(data=penguins, x='flipper_length_mm', y='body_mass_g', color='green', scatter=True)
```

### Line Plot

```python
sns.lineplot(data=penguins, x='flipper_length_mm', y="body_mass_g", hue="island", style='sex', palette="Set1")
```

### Joinplot

```python
sns.jointplot(data=penguins, x='flipper_length_mm', y="body_mass_g", kind='reg')
```

### Barplot

```python
sns.set_context('paper')
sns.barplot(data=penguins, x='species', y='flipper_length_mm', hue='island', estimator=np.mean)
```

### Count Plot

```python
sns.countplot(data=penguins,x='species', palette='Dark2', hue='island')
```

### Box Plot

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g', hue='island')
```
