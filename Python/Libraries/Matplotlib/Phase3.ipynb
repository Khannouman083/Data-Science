## Multivariate Analysis

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
```

```python
data = {
    "Salary" : [50000, 60000, 70000, 80000, 110000, 115000, 150000 ,35000, 40000],
    "Dept"   : ["HR", "IT", "Finance", "HR", "HR", "IT", "Sales", "Sales", "Finance"],
    "Age" : [50, 25,45,32,76, 35,21, 34,29],
    "Experience" : [1,2,3,4,2,4,1,5,3]
}
df = pd.DataFrame(data)
df
```

### Bubble Plot (3 Numerical Columns)

```python
plt.scatter(df["Age"], df["Salary"], s=df["Experience"]*10, color= "Blue")
plt.minorticks_on()
plt.grid()
plt.show()
```

### Bubble Plot (2 Numerical and 1 Categorical Columns)

```python
color_map = {"HR": "red", "IT": "green", "Finance": "blue", "Sales": "purple"}
plt.scatter(df["Age"], df["Salary"], c=df["Dept"].map(color_map))
handles = [mpatches.Patch(color=color, label=dept) for dept, color in color_map.items()]
plt.legend(handles=handles, title="Department")
plt.show()
```

```python
for dept, color in color_map.items():
    df_dept = df[df["Dept"] == dept]
    plt.scatter(df_dept["Age"], df_dept["Salary"], color=color, label=dept)
plt.grid()
plt.minorticks_on()
plt.legend()
```

## Object Oriented API

```python
fig, axs = plt.subplots(1,3, figsize=(15, 7))

# line Plot
axs[0].plot(df["Age"], marker = 'o', linestyle='--', color='red', linewidth=2, label= ["Age"])
axs[0].legend()
axs[0].set_title("Line Plot")
axs[0].grid()

# Historgram
axs[1].hist(df["Salary"], bins=5, color="green", edgecolor='black')
axs[1].legend()
axs[1].set_title("Histogram")
axs[1].grid()

# Boxplot
axs[2].boxplot([df["Salary"], df["Age"]])
axs[2].set_title("Boxplot")
axs[2].grid()
plt.savefig("Figure.png")
plt.show()
```

### Multiple Plots in the same plot

```python
data2 ={
    "Years": [2022,2023,2024,2025],
    "Sales": [100, 150, 200, 250],
    "Profit": [20, 30, 50, 70] , 
    "Expenses": [50, 70, 90, 110]
}

df2= pd.DataFrame(data2)
df2
```

```python
plt.plot(df2["Years"], df2["Sales"], label="Sales")
plt.plot(df2["Years"], df2["Profit"], label="Profit")
plt.plot(df2["Years"], df2["Expenses"], label="Expenses")

plt.title("Multiple Plots")
plt.xlabel("Years")
plt.ylabel("Values")
plt.legend()
plt.grid()
plt.show()
```

### 3D plots 

```python
ax= plt.axes(projection='3d')
ax.plot(df["Age"], df["Salary"], df["Experience"])
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")
plt.title("3D Scatter Plot")
plt.show()
```

### Another Method for 3D plots

```python
import plotly.express as px
import plotly.io as pio
# Set renderer to open in browser
# pio.renderers.default = "browser"
fig= px.scatter_3d(df, x="Age", y="Salary", z="Experience", title="3D Scatter Plot with plotly")
fig.show()
```
