## Bivariate Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Salary" : [50000, 60000, 70000, 80000, 110000, 115000, 150000 ,35000, 40000],
    "Dept"   : ["HR", "IT", "Finance", "HR", "HR", "IT", "Sales", "Sales", "Finance"],
    "Age" : [50, 25,45,32,76, 35,21, 34,29]
}
```

```python
df = pd.DataFrame(data)
df
```

### Scatter Plot

```python
plt.scatter(df["Age"], df["Salary"], color="cyan")
plt.grid()
plt.show()
```

### Line Plot

```python
sorted_age = df["Age"].sort_values()
plt.plot(sorted_age, df["Salary"],color = "red", marker= "o",linestyle = "--", linewidth = 2)
plt.show()
```

### Bar Chart

```python
plt.bar(sorted_age, df["Salary"], color='green', width=5, edgecolor= 'black')
plt.show()
```

### Bivariate (Numerical - Categorical)

```python
hr_sal = df[df["Dept"] == "HR"]["Salary"]
it_sal = df[df["Dept"] == "IT"]["Salary"]
finance_sal = df[df["Dept"] == "Finance"]["Salary"]
Sales_sal = df[df["Dept"] == "Sales"]["Salary"]
```

```python
plt.boxplot([hr_sal, it_sal, finance_sal, Sales_sal])
plt.show()
```

### Pie Chart

```python
salary_by_dept = df.groupby("Dept")["Salary"].sum()
plt.pie(salary_by_dept, labels=salary_by_dept.index, autopct='%1.2f')
```

### Bar Chart

```python
hr_mean = df[df["Dept"] == "HR"]["Salary"].mean()
it_mean = df[df["Dept"] == "IT"]["Salary"].mean()
finance_mean = df[df["Dept"] == "Finance"]["Salary"].mean()
sales_mean = df[df["Dept"] == "Sales"]["Salary"].mean()
```

```python
plt.bar(["HR", "IT", "Finance", "Sales"], [hr_mean, it_mean, finance_mean, sales_mean], color= ['red', 'blue', 'green', 'orange'])
plt.show()
```
